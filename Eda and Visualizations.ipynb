```python
data1 = df['cogs']
mean=np.mean(data1)
std=np.std(data1)
print(mean,std)

outlier_u = [i for i in data1 if i>mean+3*std]
outlier_l = [i for i in data1 if i<mean-3*std]

print('Upper Outliers: ',outlier_u)
print('Lower Outliers: ',outlier_l)
```

```python
data2 = sorted(df['cogs'])
q1 = np.percentile(data2,25)
q3 = np.percentile(data2,75)
iqr = q3-q1
print(q1,q3)

outlier_u = [i for i in data2 if i>q3+1.5*iqr]
outlier_l = [i for i in data2 if i<q1-1.5*iqr]

print('Upper Outliers: ',outlier_u)
print('Lower Outliers: ',outlier_l)
```

```python
x=df[['Unit price','cogs']]
from sklearn.ensemble import IsolationForest
clf=IsolationForest(n_estimators=100,contamination=0.1,random_state=42)
clf.fit(x)
y_pred=clf.predict(x)
import matplotlib.pyplot as plt
plt.scatter(x.iloc[:,0], x.iloc[:,1], c=y_pred, cmap='viridis')
plt.title("Isolation forest outlier detection on unit price and cost")
plt.xlabel("unit price")
plt.ylabel("cost")
plt.show()
```

```python
x=df[['Unit price','cogs']]
from sklearn.neighbors import LocalOutlierFactor #removes local outliers by comparing with neighbors
clf=LocalOutlierFactor(n_neighbors=20,contamination=0.1)
y_pred=clf.fit_predict(x)

plt.scatter(x.iloc[:,0], x.iloc[:,1], c=y_pred, cmap='viridis')
plt.title("Local outlier detection on unit price and cost")
plt.xlabel("unit price")
plt.ylabel("cost")
plt.show()
```

```python
#removing outliers

#log method
df['cogs_new1']=np.log(df['cogs'])
sns.distplot(df['cogs_new1'],kde=True)
```

```python
#clipping method
percentiles = df['cogs'].quantile([0.01,0.99]).values
print(percentiles)
print(df['cogs'].mean(),df['cogs'].max())
df['cogs_clip'] = np.clip(df['cogs'],percentiles[0],percentiles[1])
print(df['cogs_clip'].mean(),df['cogs_clip'].max())
```
